```python
x=df.iloc[:,:-1]
x.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.impute import SimpleImputer
```

```python
numeric_processor = Pipeline(
    steps=[
        ('imputation_mean', SimpleImputer(missing_values=np.nan, strategy='mean')),
        ('scaler', StandardScaler())
    ]
)
```

```python
numeric_processor
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_processor = Pipeline(
    steps=[
        ('imputation_constant', SimpleImputer(strategy='constant')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ]
)
```

```python
categorical_processor
```

```python
## combine processing techniques
from sklearn.compose import ColumnTransformer
```

```python
df.info()
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('categorical', categorical_processor, ['type']),
        ('numeric', numeric_processor, [
            'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
            'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
            'pH', 'sulphates', 'alcohol'
        ])
    ]
)
```

```python
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe= make_pipeline(preprocessor,LogisticRegression())
```

```python
pipe
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn import metrics
cnf_mx = metrics.confusion_matrix(y_test, y_pred)
cnf_mx
```

```python
sns.heatmap(cnf_mx,annot=True,annot_kws = {'size':15},fmt=".0f")
plt.xlabel("Predict")
plt.ylabel("Actual")
```

```python
from sklearn.decomposition import PCA

# Applying PCA without dimensionality reduction for comparison
pipe_no_pca = make_pipeline(
    preprocessor,
    LogisticRegression(max_iter=1000)
)

# Fitting the model without PCA
pipe_no_pca.fit(X_train, y_train)

# Predictions without PCA
y_pred_no_pca = pipe_no_pca.predict(X_test)

# Confusion Matrix without PCA
conf_matrix_no_pca = metrics.confusion_matrix(y_test, y_pred_no_pca)
print("Confusion Matrix without PCA:")
print(conf_matrix_no_pca)

# ROC-AUC Curve without PCA
y_proba_no_pca = pipe_no_pca.predict_proba(X_test)
roc_auc_no_pca = roc_auc_score(y_test, y_proba_no_pca, multi_class='ovr')
print(f'ROC-AUC without PCA: {roc_auc_no_pca:.2f}')

# Applying PCA
pipe_pca = make_pipeline(
    preprocessor,
    PCA(n_components=5),
    LogisticRegression(max_iter=1000)
)

# Fitting the model with PCA
pipe_pca.fit(X_train, y_train)

# Predictions with PCA
y_pred_pca = pipe_pca.predict(X_test)

# Confusion Matrix with PCA
conf_matrix_pca = metrics.confusion_matrix(y_test, y_pred_pca)
print("Confusion Matrix with PCA:")
print(conf_matrix_pca)

# ROC-AUC Curve with PCA
y_proba_pca = pipe_pca.predict_proba(X_test)
roc_auc_pca = roc_auc_score(y_test, y_proba_pca, multi_class='ovr')
print(f'ROC-AUC with PCA: {roc_auc_pca:.2f}')
```
