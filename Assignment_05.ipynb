```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('XSS_dataset.csv')
df
```

```python
df.info()
```

```python
df.drop(df.columns[df.columns.str.contains(
    'unnamed', case=False)], axis=1, inplace=True)
```

```python
df.head(15)
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import re
ps = PorterStemmer()
nltk.download('stopwords')
```

```python
corpus = []
for i in range(len(df)):
  review = re.sub('[^a-zA-Z]',' ', df['Sentence'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
  review = ' '.join(review)
  corpus.append(review)
corpus[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
X_arr = X.toarray()
X_arr
```

```python
# visualizing as a dataframe
df_1 = pd.DataFrame(data = X_arr, columns = vectorizer.get_feature_names_out())
df_1
```

```python
y = pd.get_dummies(df['Label'])
y = y.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_arr, y, test_size = 0.2, random_state = 7)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model = model.fit(X_train, y_train)
pred = model.predict(X_test)
pred
```

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test, pred)
cf
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, pred)
accuracy
```
